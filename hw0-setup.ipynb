```python
def plot_sine(freq, dur, sr, output_path):
  samples = dur * sr
  y = np.sin(2 * np.pi * freq * np.arange(samples) / sr)
  plt.plot(np.arange(samples) / sr, y)
  plt.xlabel('Time (s)')
  plt.savefig(output_path)
```
